```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
# saved the table to a csv and uploaded to pandas
pc = pd.read_csv('info.csv')
pc.head(10)
```

```python
# Removing "Not assigned" from "Borough" columns and resetting the indices
pc = pc[pc.Borough != 'Not assigned']
pc = pc.reset_index()
pc.head(10)
```

```python
# Dropping the unnecessary old indices 
pc = pc.drop('index',axis=1)
pc.head(10)
```

```python
# replacing the only "Not assigned" in the Neighbourhood column
pc.replace(to_replace='Not assigned', value="Queen's Park", inplace=True)
pc.head(10)
```
